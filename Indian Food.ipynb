# Kaggle dataset - Indian Food
Let's explore this dataset:

Firstly I'll import the pandas library which is required to read the dataset:

```python
import pandas as pd
```

Now let's check what this dataset is about:

```python
food = pd.read_csv('indian_food.csv')
```

```python
# Checking the first few rows of the dataset:
food.head()
```

Okay so we have 9 columns, viz;
- name : name of the dish
- ingredients : main ingredients used
- diet : type of diet - either vegetarian or non vegetarian
- prep_time : preparation time
- cook_time : cooking time
- flavor_profile : flavor profile includes whether the dish is spicy, sweet, bitter, etc
- course : course of meal - starter, main course, dessert, etc
- state : state where the dish is famous or is originated
- region : region where the state belongs

I'll now check the total number of rows in this dataset:

```python
food.shape
```

Alright so we have 255 rows here. I'll now check if there's any missing value in the dataset:

```python
food.isnull().sum()
```

Hmm, only one missing value. Well, nopes! Actually the dataset description on kaggle says missing values are represented by -1 here. But firstly I'll deal with this one missing value on the region column. Since it's a categorical value, I'll fill it up with mode(most common value of this particular column):

```python
food['region'].fillna(food['region'].mode,inplace=True)
```

```python
# Let me confirm if it's filled up or not:
food.isnull().sum()
```

Alright, now I'll search for all the descriptive missing values (-1) in the dataset:

```python
for name in food['prep_time']:
    if name== -1:
        print(True)

else:

    print(False)
```

Okay, so we have many -1s in 'prep_time' column. Let's find all the locations of this column where -1 is present:

```python
food.loc[food['prep_time'].isin([-1])]
```

```python
# I need to check the statistics of prep_time and cook_time before deciding how to fill their values:
food.describe()
```

```python
#Let me now check the spread in prep_time:
food['prep_time'].value_counts()
```

Okay, so 98 dishes have a prep_time of 10 minutes, 40 dishes have prep_time of 20 minutes, 22 dishes have prep_time of 15 minutes and so on. Plus, the maximum values of both prep_time & cook_time are pretty high. So I can treat them as continuous variables. Firstly, I'll deal with prep_time. Replacing all -1 with the mean value:

```python
food['prep_time'].replace([-1], food['prep_time'].mean() ,inplace=True)
```

I'll now check for -1 in 'cook_time' column/variable:

```python
food.loc[food['cook_time'].isin([-1])]
```

Now, as cook_time is also a continuous variable, I'll replace all -1 values of this column too with the mean:

```python
food['cook_time'].replace([-1], food['cook_time'].mean() ,inplace=True)
```

Next, we have 'flavor_profile', a categorical variable. So I'll replace its -1 values with mode:

```python
food['flavor_profile'].replace([-1], food['flavor_profile'].mode ,inplace=True)
```

Let me have a look at the bottom 20 rows now:

```python
food.tail(20)
```

Filling missing values of flavor_profile with mode:

```python
food['flavor_profile'].replace(['-1'], food['flavor_profile'].mode() ,inplace=True)
```

Similarly, as 'state' and 'region' variables are also categorical, I'm filling missing values of these too with mode:

```python
food['state'].replace(['-1'], food['state'].mode() ,inplace=True)
food['region'].replace(['-1'], food['region'].mode() ,inplace=True)
```

Let me take one more look at bottom 20 rows:

```python
food.tail(20)
```

Now the data is all cleaned. I'll now visualize the data with the help of plots. But first I need to import numpy, matplotlib and seaborn libraries for this purpose:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

Let me first see how many dishes are vegetarian and how many are non-vegetarian:

```python
food['diet'].value_counts().plot.bar()
```

Clearly there are around 25 non-vegetarian dishes, and on the other hand, around 230 vegearian dishes.

Now let's take a look at state-wise 'diet' distribution:

```python
pd.crosstab(food['state'],food['diet'])
```

So in Andhra Pradesh state there's no non-vegetarian dish but 10 vegetarian dishes. Similarly in Assam state, there are 10 non-vegetarian dishes and 11 vegetarian dishes. And so on.

```python
# Let's check how many types of flavor profiles are there:
food['flavor_profile'].value_counts().plot.bar()
```

Okay so most dishes, around 160, are spicy. Over 80 dishes are sweet. A little less than 10 are bitter & very few are sour.

```python
# Seeing this in pie chart will be fun:
labels = ['spicy','sweet','bitter','sour']
sizes= [160,85,8,2]
colors = ['blue','red','green','lightskyblue']
explode= [0,0,0,0] #explode 1st slice

#plot
plt.pie(sizes,explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=True)
# autopoint means in what format u want stuff..1.1f means floating point, shadow gives shadow to pizza structure(c yellow wala)

plt.axis('equal')
plt.show()
```

Let's now take a look at the 'course' column:

```python
food['course'].value_counts().plot.bar()
```

This shows most of the dishes, over 120, are of "main course" type. Second highest type is "dessert" which is about 80. Least registered dish type in the dataset is of "starter" dish type.
